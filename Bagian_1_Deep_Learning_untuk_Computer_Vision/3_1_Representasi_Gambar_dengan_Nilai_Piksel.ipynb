```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Membaca gambar menggunakan OpenCV
img = cv2.imread('fruit.jpeg', cv2.IMREAD_GRAYSCALE)

# Resize gambar untuk visualisasi lebih sederhana (opsional)
img_resized = cv2.resize(img, (20, 20), interpolation=cv2.INTER_NEAREST)

# Menyiapkan figure
fig, axes = plt.subplots(1, 2, figsize=(14, 8))

# Visualisasi kiri: Gambar dengan piksel diperbesar
axes[0].imshow(img_resized, cmap='gray', interpolation='nearest')
axes[0].set_title("Pixelated Image")
axes[0].axis('off')

# Visualisasi kanan: Gambar dengan nilai intensitas
axes[1].imshow(img_resized, cmap='gray', interpolation='nearest')
axes[1].set_title("Image with Pixel Values")

# Menambahkan nilai piksel ke gambar
for i in range(img_resized.shape[0]):
    for j in range(img_resized.shape[1]):
        pixel_value = img_resized[i, j]
        axes[1].text(j, i, str(pixel_value),
                     ha='center', va='center',
                     color='white' if pixel_value < 128 else 'black')

plt.tight_layout()
plt.show()
```
