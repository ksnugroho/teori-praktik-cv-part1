```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('fruit.jpeg')
```

# Konversi Gambar Grayscale menjadi Array Terstruktur

```python
# Mengonversi format dari RGB ke Grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
# Resize gambar ke 25 × 25 piksel
img_gray_small = cv2.resize(img_gray,(25,25))

plt.imshow(img_gray_small, cmap='gray')
plt.show()
```

```python
print(img_gray_small)
```

```python
# Menampilkan nilai piksel 4 baris pertama
print("Nilai Piksel (4 Baris Pertama):")
print(img_gray_small[:4])
```

# Konversi Gambar Berwarna (RGB) menjadi Array Terstruktur

```python
# Mengonversi format dari BGR (default OpenCV) ke RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
print(f"Dimensi Gambar: {img.shape}")  # Output: (height, width, channels)
```

```python
# Mengambil bagian bawah kanan 3 × 3 piksel
crop = img[-3:, -3:, :]
```

```python
# Menampilkan potongan gambar
plt.imshow(crop)
plt.show()

# Menampilkan nilai piksel RGB
print("Nilai RGB dari Bagian 3 × 3:")
print(crop)
```
